#### remove '+' sign attached to the values in the age column

```python
age_map = {  '18+':18,
             '7+':7,
             '13+':13,
             'All':0,
             '16+':16
            }
```

```python
df['AgeCopy'] = df['Age'].map(age_map)
```

```python
df['new_Rotten_Tomatoes'] = df['Rotten Tomatoes']
```

```python
df['new_Rotten_Tomatoes'] = (
    df['new_Rotten_Tomatoes'].str.replace('/100', '', regex=False)
     .str.strip()
     .astype('Int64')   # capital I â†’ Nullable Int dtype
)
```
